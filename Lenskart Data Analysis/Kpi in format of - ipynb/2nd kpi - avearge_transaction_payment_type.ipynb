```python
import pandas as pd

# Read data from CSV file
revenue_data_df = pd.read_csv("revenue_data.csv")

# Calculate average transaction revenue
average_transaction_revenue = revenue_data_df.groupby(['store_id', 'payment_method'])['revenue'].mean().reset_index()

# Reorder columns
average_transaction_revenue = average_transaction_revenue[['store_id', 'payment_method', 'revenue']]


display(average_transaction_revenue)
average_transaction_revenue.to_csv("average_transaction_payment_type.csv", index=False)
```
